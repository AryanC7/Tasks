### Bin the age column. Create a new column 'age_bins' with the Age values binned as [0, 20, 40, 60, 80, 100]

### Make a chart showcasing number of male and female patients in the given data, explain why the data has more values of one gender over the other

```python
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.countplot(data=df, x="Gender", palette="Set1")

plt.xlabel("Gender")
plt.ylabel("Number of Patients")
plt.title("Distribution of Patients by Gender")

plt.show()
```

* Why are female patients more in number than males?
  * *Breast Tissue Presence:*
    * Breast tissue contains structures like milk ducts and lobules that are susceptible to cancerous growth. The higher volume of breast tissue increases the chances of breast cancer occurrence.
  * *Hormonal Influence:*
    * Female sex hormones, primarily estrogen and progesterone, play a significant role in breast development and maintenance. These hormones can also influence the growth of certain types of breast cancer.
  * *Mammography and Diagnosis:*
    * Mammography, a common method of breast cancer screening, is more effective in detecting breast cancer in females due to their larger breast tissue volume. This results in more breast cancer cases being diagnosed in females.

### 4. ER, PR, HER2 status

```python
df["ER status"] = df["ER status"].map({"Positive": 1, "Negative": 0})
df["PR status"] = df["PR status"].map({"Positive": 1, "Negative": 0})
df["HER2 status"] = df["HER2 status"].map({"Positive": 1, "Negative": 0})
```

```python
df.head()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(6, 6))
sns.countplot(data=df, x="ER status", hue="Target", palette="Set1")
plt.title("Distribution of ER Status")
plt.xlabel("ER Status")
plt.ylabel("Count")
plt.xticks([-0.2, 0.2], ["Negative", "Positive"])
plt.show()

plt.figure(figsize=(6, 6))
sns.countplot(data=df, x="PR status", hue="Target", palette="Set1")
plt.title("Distribution of PR Status")
plt.xlabel("PR Status")
plt.ylabel("Count")
plt.xticks([-0.2, 0.2], ["Negative", "Positive"])
plt.show()

plt.figure(figsize=(6, 6))
sns.countplot(data=df, x="HER2 status", palette="Set1")
plt.title("Count of HER2 Status")
plt.xlabel("HER2 Status")
plt.ylabel("Count")
plt.xticks([-0.05, 1], ["Negative", "Positive"])
plt.show()
```

**Hormone Receptor(HR)** - *Positive*
  * breast cancer cells have either estrogen (ER) or progesterone (PR) receptors or both and cancer cells with both of these as positive, can be treated with hormone therapy drugs that lower estrogen levels or block estrogen receptors.
  * they tend to grow slowly.
  
**Hormone Receptor(HR)** - *Negative*
  * breast cancers have no estrogen or progesterone receptors and treatment with hormone therapy drugs is not helpful for these cancers.
  * they tend to faster than HR positive cells.
  
**Triple-negative**
  * breast cancer cells don’t have estrogen or progesterone receptors and also don’t make any or too much of the protein called HER2.
  * triple-negative breast cancers grow and spread faster than most other types of breast cancer.
  
**Triple-positive**
  * cancers are ER-positive, PR-positive, and HER2-positive.
  * these cancers can be treated with hormone drugs as well as drugs that target HER2.
  
**Inference**
  * if a patient has more positive results in the ER, PR and HER2 tests, then the chances of survival increases and less invasive/aggressive drugs have to be used on them to treat the cancer cells.
  * most of the patients in our dataset don't have triple positive but have HR - Positive which means their treatment can still be made easier and less painful.
